```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from google.colab import files
from google.colab import files
from google.colab import files
uploaded = files.upload()
```

```python
img = cv2.imread(list(uploaded.keys())[0])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for plotting
plt.imshow(img)
plt.title('/content/cows eye.jpg')
plt.axis('off')
plt.show()
```

```python
def kmeans_color_quantization(image, k):
    # Reshape image to (num_pixels, 3)
    data = image.reshape((-1, 3))
    data = np.float32(data)

    # Apply KMeans
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(data)
    centers = np.uint8(kmeans.cluster_centers_)

    # Replace each pixel by its center
    quantized = centers[labels].reshape(image.shape)
    return quantized

# k=3
k3_result = kmeans_color_quantization(img, 3)
# k=5 (your choice)
k5_result = kmeans_color_quantization(img, 5)

# Plot results
fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].imshow(img)
axs[0].set_title("Original")
axs[1].imshow(k3_result)
axs[1].set_title("K-means k=3")
axs[2].imshow(k5_result)
axs[2].set_title("K-means k=5")
for ax in axs:
    ax.axis('off')
plt.show()
```

```python
#Crop image to 100x100 for agglomerative clustering
cropped = cv2.resize(img, (100, 100))
plt.imshow(cropped)
plt.title('Cropped 100x100')
plt.axis('off')
plt.show()
```
